## Imports

```python
from pathlib import Path
import matplotlib.pyplot as plt
import polars as pl
import numpy as np
```

## Loading data

```python
data_path = Path("../data/data.csv")

df = pl.read_csv(data_path)
```

## Visualization

```python
xs = df["x"].to_numpy()
ys = df["y"].to_numpy()
states = df["state"].to_numpy()

width = 1000
height = 1000

arr = np.zeros((height, width), dtype=int)

for x, y, s in zip(xs, ys, states):
    arr[y, x] = int(s)
```

```python
# arr[arr != 0] = 1
```

```python
plt.figure(figsize=(10, 10))

plt.imshow(arr, cmap="jet")

# plt.xlim([200, 800])
# plt.ylim([200, 500])

plt.axis("off")

plt.tight_layout()

plt.savefig(
    "../figures/20260102 - Diffusion Limited Aggregation - colorful picture.png",
    dpi=300,
)
```
